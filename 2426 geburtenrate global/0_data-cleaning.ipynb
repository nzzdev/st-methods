```python
import pandas as pd
```

### List of regions

```python
ctr = pd.read_csv('data/ctr.csv').rename({"Economy": "Location"}, axis=1)
ctr
```

### Live births

```python
d1 = pd.read_csv("data/IHME_GBD_2021_FERTILITY_1950_2100_LB_Y2024M03D15.CSV")
d1['year'] = pd.to_datetime(d1['year_id'], format="%Y") # convert year_id to datetime
d1 = d1.drop_duplicates(subset=['location_id', 'year_id', 'scenario_id'], keep='first') ## Somehow the year 2022 is twice in there
d1.rename({
    'location_name': 'Location',
    'scenario_name': 'Scenario',
}, axis=1, inplace=True)
d1 = pd.merge(d1, ctr, on='Location', how='left')
d1.query("Location == 'Switzerland'")
```

```python
d1[['Location', 'Code', 'Region', 'year', 'Scenario', 'val', 'lower', 'upper']].to_pickle("cleaned/births.pkl")
```

### Total fertility rate

```python
d2 = pd.read_csv("data/IHME_GBD_2021_FERTILITY_1950_2100_TFR_Y2024M03D15.CSV")
d2['year'] = pd.to_datetime(d2.year_id, format="%Y")
d2.rename({
    'location_name': 'Location',
    'scenario_name': 'Scenario',
}, axis=1, inplace=True)
d2 = pd.merge(d2, ctr, on='Location', how='left')
d2
```

```python
d2[['Location', 'Code', 'Region', 'year', 'Scenario', 'val', 'lower', 'upper']].to_pickle("cleaned/tfr.pkl")
```
