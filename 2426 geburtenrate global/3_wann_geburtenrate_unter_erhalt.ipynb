```python
d1 = pd.read_pickle('cleaned/tfr.pkl')
d1 = d1.query("Scenario == 'Reference' & Code.isna() == False")
d1
```

```python
d1.query("year == 2100 & val >= 2.1")
```

```python
d2 = d1.query("val < 2.1").groupby('Location').first().reset_index()
d2
```

```python
nzz = pd.read_csv("data/nzz-countries.csv")
nzz.head()
```

```python
d3 = pd.merge(nzz, d2, left_on='Weltbank', right_on='Location', how='left')
d3.query("val.isna() == True")
```

```python
d3['Jahr'] = d3.year.dt.year
d3[['NZZ', 'Jahr']]
```

```python
d3[['NZZ', 'Jahr']].to_csv('out/weltkarte.csv', index=False)
```
