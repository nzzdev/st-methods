```python
import pandas as pd
import glob
import altair as alt
```

```python
GDT=pd.read_csv("GDT_database.csv")
print(len(GDT))
```

```python
GDT
```

```python
GDTclean=GDT[(GDT["crit1"]&GDT["crit2"]&GDT["crit3"]==1)]
GDTclean=GDTclean[GDT["doubtterr"]!=1]

print(len(GDTclean))
GDTclean
```

```python
GDTclean["incidentno"]=1
GDTclean["ncasualties"]=GDTclean["nkill"]+GDTclean["nwound"]
GDTclean
```

```python
GDTcountry=GDTclean.groupby(["country_txt"]).sum()
GDTcountry
```

```python
GDTcountry=GDTcountry.sort_values(by="incidentno", ascending=False)
GDTcountry
```

```python
pd.options.display.max_columns=100
pd.options.display.max_rows=10
GDTcountry
```

```python
GDTcountry=GDTcountry.reset_index()
GDTcountry
```

```python
GDT
```

```python
from altair import *
import altair as alt
Chart(GDTcountry.head(20)).mark_bar(color='#292fb3').encode(
    x=alt.X('ncasualties:Q',axis=alt.Axis(title='Tote und Verletzte',format='d')),
    y=alt.Y('country_txt', axis=alt.Axis(title='L√§nder'),
            #we then sort age according to the field "people" by summing, and in descending order
            sort=alt.SortField(field='ncasualties', order='descending', op="mean")),
   
)
```

```python
Kill_Casualties=[]
Kill_KillORInj=[]
Inj_Casualties=[]
Inj_KillORInj=[]

for index, row in yearbased.iterrows():
   
        Kill_Casualties.append(row["nwound"])
        Kill_KillORInj.append("Wound")

        Inj_Casualties.append(row["nkill"])
        Inj_KillORInj.append("Kill")
    
print(len(Kill_Casualties),len(Kill_KillORInj),len(Inj_Casualties),len(Inj_KillORInj))
```

```python
alt.Chart(yearbased).mark_bar( barSize=12, color='#292fb3').encode(
        alt.Y('iyear', axis=alt.Axis(format='d', title='year'), 
                                       scale=alt.Scale(domain=(1969, 2016))),
    

    x='incidentno').configure_cell(height=400, width=800).configure_scale(bandSize=600)
```

```python
KillDf=pd.DataFrame()
KillDf['casualties'] = Kill_Casualties
KillDf['KillORInj'] = Kill_KillORInj
KillDf=pd.concat([yearbased,KillDf], axis=1)
InjDf=pd.DataFrame()
InjDf['casualties'] = Inj_Casualties
InjDf['KillORInj'] = Inj_KillORInj
InjDf=pd.concat([yearbased,InjDf], axis=1)
```

```python
KillANDInj=pd.concat([KillDf,InjDf])
len(KillANDInj)
```

```python

from altair import *

Chart(
    KillANDInj
).mark_bar( barSize=12).encode(
    alt.X('iyear', axis=alt.Axis(format='d', title='year'), 
                                       scale=alt.Scale(domain=(1969, 2016))),
    color=Color("KillORInj:N",
        legend=Legend(
            title='Injured Or Killed',
        ),
        scale=Scale(
            domain=['Kill', 'Wound'],
            range=['#fae1e0', '#b39829'],
        ),
    ),
    
    y='casualties',
).configure_cell(height=400, width=800).configure_scale(bandSize=600)
```
