```python
import pandas as pd
import pycountry
```

```python
py_countries = pd.read_csv('py_countries.csv')
q_names = pd.read_csv('country_ids.csv')
q_iso = pd.read_csv('country_iso3.csv')
```

```python
py_countries['iso_2'] = py_countries['iso_2'].str.replace("_", "")
```

```python
name = 'Netherlands Antilles'
pycountry.countries.get(name = name)
```

```python
# return iso3
errors = []

def get_iso3(iso2):
    try: 
        return pycountry.countries.get(alpha_2 = iso2).alpha_3
    except:
        errors.append(iso2)
```

```python
py_countries['iso_3'] = py_countries['iso_2'].apply(get_iso3)
```

```python
py_countries[py_countries['iso_2'].isin(errors)]
```

```python
notfound = []
for py in py_countries['iso_3'].to_list():
    if py not in q_iso['ISO3'].to_list():
        notfound.append(py)
```
