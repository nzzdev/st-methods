Loading the libaries

```python
import pandas as pd
import selenium
import time
import sys
```

Setting up selenium for going over the transfermarkt page and clicking through all the pages

```python
url="https://www.transfermarkt.de/spieler-statistik/marktwertnachwm/marktwertetop?land_id=0&spielerposition_id=0&altersklasse=&plus=1"
from selenium import webdriver
from selenium.webdriver.common import action_chains, keys

driver = webdriver.Chrome()
#driver = webdriver.Firefox()
driver.get(url)
WMarray=[]
```

This is the function that gets called up in a for-loop to scrape the Transfermarkt page and click through the subsites that are Javascript-based

```python
#first there is an empty array for all the data
WMarray=[]

#this is the function that gets called up in a for loop
def wmfunction():

    #unfortunately the structure of the webpage is a little bit chaotic. It is rows filled with cells that
    #contain 2 rows that again contain cells and especially the second one tends to be popualted in a non-coherent pattern,
    #which makes the whole iteration rather complicated
    #For the outer-rows, we define one index, that always gets set to zero, 
    #when a new page is opened by clicking on the "next page" button
    #on each page, we find 25 players, so we let this counter count to 25 and then click on the next page
    a=0
    
    #we get all outer-rows
    rows = driver.find_elements_by_tag_name("tr")
    for row in rows:
        #and the outer-cells in this row
        cells=row.find_elements_by_tag_name("td")
        
        #for each of these cells, 
        for cell in cells:
                #we again find row elements
                innerrows = driver.find_elements_by_tag_name("tr")
                
                #through which we iterate. Here we use the second index counter, that keeps track of the cell number
                #and gets reset to zero when we get into a new row
                for innerrow in innerrows:
                    i=0
                    
                    #we get the innercells
                    innercells=innerrow.find_elements_by_tag_name("td")
                    #and then iterate through these innercells where we can finally do the real scraping
                    for innercell in innercells:
                        i+=1
                       
                          
                    #if our index is at 4, we retrieve the players name 
                        if i==4:
                            playername=innercell.text
                            #and additionally also try to get the link to the player page and his unique id in the transfermarkt sysrem
                            try:
                                playerattributes=innercell.find_elements_by_tag_name("a")
                                PlayerHref=playerattributes[0].get_attribute("href")
                                PlayerId=playerattributes[0].get_attribute("id")
                            except IndexError:
                                PlayerHref=""
                                PlayerId=""
                                

                            print(playername, PlayerHref, PlayerId)
                        #if our index is 5, we retrieve the players position
                        if i==5:
                            position=innercell.text
                        #if our index is 6, we retrieve the players age
                        if i==6:
                            alter=innercell.text
                         #if our index is 7, we try to retrieve the player's country, his country's page at Transfermarkt and his country's unique id
                        if i==7:
                            try:
                                flaggediv=innercell.find_elements_by_tag_name("a")
                                flaggeinnerdiv=flaggediv[0].find_elements_by_tag_name("img")
                                Land=(flaggeinnerdiv[0].get_attribute("alt"))
                                LandHref=(flaggediv[0].get_attribute("href"))
                                LandId=(flaggediv[0].get_attribute("id"))
                            except IndexError:
                                Land=""
                                LandHref=""
                                LandId=""
                        #if our index is 8, we try to retrieve the player's club where he played before the world cup, his club's page at Transfermarkt and his club's unique id
                        if i==8:
                            try:
                                PreWMclubdiv=innercell.find_elements_by_tag_name("a")
                                PreWMclubinnerdiv=PreWMclubdiv[0].find_elements_by_tag_name("img")
                                PreWMClub=(PreWMclubinnerdiv[0].get_attribute("alt"))
                                PreWMClubHref=(PreWMclubdiv[0].get_attribute("href"))
                                PreWMClubId=(PreWMclubdiv[0].get_attribute("id"))
                            except IndexError:
                                PreWMClub=""
                                PreWMClubHref=""
                                PreWMClubId=""
                        #if our index is 9, we retrieve his value pre-WorldCup
                        if i==9:
                            VorWMWert=innercell.text
                        #if our index is 10, we retrieve his value post-WorldCup
                        if i==10:
                            PostWMWert=innercell.text
                        #if our index is 11, we retrieve the difference in value
                        if i==11:
                            WMDiff=innercell.text
                        #if our index is 12, we try to retrieve the player's club where he played after the world cup, his club's page at Transfermarkt and his club's unique id
                        if i==12:
                            try:
                                clubdiv=innercell.find_elements_by_tag_name("a")
                                clubinnerdiv=clubdiv[0].find_elements_by_tag_name("img")
                                Club=(clubinnerdiv[0].get_attribute("alt"))
                                ClubHref=(clubdiv[0].get_attribute("href"))
                                ClubId=(clubdiv[0].get_attribute("id"))
                            except IndexError:
                                Club=""
                                ClubHref=""
                                ClubId=""
                        #if our index is 13, we retrieve the player's current market value
                        if i==13:
                            Wertjetzt=innercell.text
                        #if our index is 14, we retrieve how the player's market value changed in percent between the world cup and now
                        if i==14:
                            WertdiffjetztProzent=innercell.text
                        #if our index is 14, we retrieve how the player's market value changed between the world cup and now in absolute numbers
                        if i==15:
                            Wertdiffjetzt=innercell.text
                            #finally we push all these variables into an array that gets added to the outer-array
                            Spielerarray=[playername,PlayerHref,PlayerId,position,alter,Land,LandHref,LandId,PreWMClub,PreWMClubHref,PreWMClubId,VorWMWert,PostWMWert,WMDiff,Club,ClubHref,ClubId,Wertjetzt,WertdiffjetztProzent,Wertdiffjetzt]
                            WMarray.append(Spielerarray)
                            a+=1
                            print("ROW:", a)
                        if a==25:
                            #and if we reached the 25th outer-row, we click to get to the next page
                            try:
                                driver.find_element_by_css_selector(".naechste-seite").click()
                                time.sleep(20)
                                a=0
                                return
                            except ElementNotVisibleException:
                                return
                            
for i in range(31):
    wmfunction()
              
```

```python
#finally we create a dataframe from this array with the right columnames, drop the duplicates and push it out as a CSV
transferDF1=pd.DataFrame(WMarray, columns=["Spielername","PlayerHref","PlayerId","Position","Alter","Land","LandHref","LandId","ClubVorWM","PreWMClubHref","PreWMClubId","VorWMWert","NachWMWert","WertDifferenz","ClubJetzt","ClubJetztHref","ClubJetztId","Wertjetzt","WertdiffjetztProzent","Wertdiffjetzt"])
transferDF_dp=transferDF1.drop_duplicates()
print(len(transferDF1))
print(len(transferDF_dp))

transferDF_dp.to_csv("Transfermarkt/Transfermarkt_WM_Werte_clean.csv")
```
