```python
from bs4 import BeautifulSoup as BS
import bs4
import re
import pandas as pd
import requests
import numpy as np
import time
from bs4 import BeautifulSoup
import requests
```

We scrape all available data from the FIFA webpage for the 2014 World Cup, for this we go through each of the statistic pages and read out the table element. The first webpage that is being scraped will be commented in detail

```python
#We create an empty array to populate with data
WMarray=[]

#We send the request to the fifa webpage with the requests library
headers = {'User-Agent': 'Mozilla/6.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
url='http://www.fifa.com/worldcup/archive/brazil2014/statistics/players/goal-scored.html'
r=requests.get(url, headers=headers)
print(url)

#we parse the response html with the beautifulsoup library
soup = BeautifulSoup(r.content, 'html.parser')

#we find all tables in this html and get the first one of them and get all rows of this table
tables=(soup.find_all('table'))
rows=(tables[0].find_all("tr"))

#we iterate through each of these rows
for row in rows:
    #and find all the cells in these rows
    cells=(row.find_all("td"))
    #also we use an index counter to count through the cells, this index counter gets reset to zero whenever we enter a new row
    i=0
    for cell in cells:
        #and this index counter gets increased by one for each cell 
        i+=1
        #now we extract the text from each cell and assign it to a variable
        #if this index is 1, this means the cell has the player's name
        if i==1:
            Playername=(cell.text)
            img=(cell.find_all("img"))
            flag=img[0].get('title')
        #if this index is 2, it is the cell containing the number of matches played
        if i==2:
            MatchesPlayed=(cell.text)
        #if this index is at 3, it is the cell containing the number of minutes played
        if i==3:
            MinutesPlayed=(cell.text)
        #if this index is at 4, it is the cell containing the goals scored
        if i==4:
            Goals=(cell.text)
        #if this index is at 5, it is the cell containing the assists played
        if i==5:
            Assists=(cell.text)
        #if this index is at 6, it is the cell containing the penalties shot
        if i==6:
            Penalties=(cell.text)
        #if this index is at 7, it is the cell containing the goals shot with the left foot
        if i==7:
            LeftyGoals=(cell.text)
        #if this index is at 8, it is the cell containing the goals shot with the right foot
        if i==8:
            RightyGoals=(cell.text)
        #if this index is at 6, it is the cell containing the goals scored by headers
        if i==9:
            HeaderGoals=(cell.text)
            #finally we add all these variables into an array
            Spielerarray=[flag,Playername,MatchesPlayed,MinutesPlayed,Goals,Assists,Penalties,LeftyGoals,RightyGoals,HeaderGoals]
            print(Spielerarray)
            #and nest this array into the WMarray that gives us the base structure for a dataframe
            WMarray.append(Spielerarray)

#finally, we create from this array a dataframe and define the columnnames of this dataframe
WMarray=pd.DataFrame(WMarray, columns=["Country","Spielername","MatchesPlayed","MinutesPlayed","Goals","Assists","Penalties","LeftyGoals","RightyGoals","HeaderGoals"])

#and make a csv from this dataframe
WMarray.to_csv("Fifa_Players/GoalsScored_players.csv", index=False)

            
            
        
```
