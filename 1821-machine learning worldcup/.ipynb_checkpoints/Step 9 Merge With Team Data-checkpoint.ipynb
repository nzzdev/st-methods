```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = 100
pd.options.display.max_rows = 400
import glob
```

Here we just merge the data of all teams onto the dataframe with WhoScored Data, Fifa Data and Transfermarkt Data

```python
allFiles = glob.glob("teamsafterfinal/*.csv")  ##'*' means any file name can be grabbed
cleanDF=pd.DataFrame()
```

```python
i=0
older=0
newer=0
for file in allFiles:
    print(i)
    print(file)

    i+=1
    newer="_"+(file[16:-13])
    print(newer)

    data=pd.read_csv(file)
    data=data.drop_duplicates()
    if i==1:
        maindf=data
        older=newer

    else:
        maindf=pd.merge(maindf,data,on=["team-link","Flag"], how="outer", suffixes=(older,newer))
        older=newer



    
```

Here is the lookup for the countries

```python
LaenderCountriesLookup=pd.read_csv("LaenderCountriesLookup.csv")
LaenderCountriesLookup
```

```python
maindf_added=pd.merge(maindf,LaenderCountriesLookup,left_on="team-link",right_on="LaenderStatsName",how="inner")
maindf_added
```

```python
Playerdata=pd.read_csv("HandCleanedMerge.csv")
Playerdata
```

```python
Player_Countries_df=pd.merge(Playerdata, maindf_added, left_on="Land",right_on="AllMergedName", how="outer", suffixes=["_player","_country"])
Player_Countries_df
```

```python
Player_Countries_df.to_csv("AllMerged_CountriesIncluded.csv",index=False)
```
