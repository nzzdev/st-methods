```python
gruende=data["dep"].unique().tolist()
gruende
```

```python
for i in gruende:
    subdata=data[data["dep"]==i]
    subdata=subdata[["dep",'1980','1985','1990','1995','2000',
 '2005',
 '2006',
 '2007',
 '2008',
 '2009',
 '2010',
 '2011',
 '2012',
 '2013',
 '2014',
 '2015',
 '2016',
 '2017',
 '2018',
 '2019',
 '2020']]
    subdata=subdata.fillna(0)
    array=subdata.groupby(by="dep").sum()
    trans_array=array.transpose()
    trans_array=trans_array.reset_index()
    trans_array=trans_array.rename(columns={"index":"Jahr"})
    chart=alt.Chart(trans_array).mark_bar().encode(
    x='Jahr:T',
    y=i,
)
    display(chart)
```
