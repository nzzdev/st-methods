# Wiki Terrorism Scraper
This is a scraper which is designed to scrape Wikipedias list of terror incidents from 1970 onwards up till now. The time frame was chosen because it is the same time frame that the University of Maryland database is using and so it is easy to compare these two against each other.

```python
# coding: utf-8
# You need to pip3 install bs4, requests
import re
import os
import csv
import time
import requests
import datetime as dt
from datetime import datetime

from bs4 import BeautifulSoup as BS
```

The core idea of all the following scrapers is the same.
1. Create a csv
2. Go to the wikipedia page and find the html table(s)
3. Go through each row of the table(s)
4. In each row, go through each column so that you get all cells
5. Write them to the csv

```python
#this scraper cover 1970 to 1979
with open(("terror_1970to1979_{}.csv").format(localtime), 'w') as resultsfile:
    
    resultsfileWriter = csv.writer(resultsfile)
    firstrow=("day","month","year","attacktype","deaths","injured","location","details","attacker","context","country","link")
    resultsfileWriter.writerow(firstrow)
    
    for year in range(1970,1980):
            year=str(year)
            link=("https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_{}".format(year))
            print(link)
            page = requests.get(link).text
            soup = BS(page, 'html.parser')
            #print(soup)
            table = soup.find('table', class_="wikitable sortable")
            #print(table)
            rows = table.find_all('tr')
            for row in rows:
                i=0
                cells=(row.find_all("td"))
                for cell in cells:
                    #Scraper uses i to count through the columns and define what is recorded in each column
                    i+=1
                    if i==1:
                        celldate=cell.text
                        datelist=celldate.split()
                        #scraper retrieves the text from the date cell and splits it into a list
                        #a new element of the list starts whenever there is space between the strings in the date cell
                        try:
                            #it tries to take the last element of this list as the day and the first one as the month
                            day=datelist[-1]
                            month=datelist[0]
                            #this checks whether by mistake the order of month and day was the wrong way around 
                            #and they have to be switched
                            if day[0].isdigit():
                                continue
                            else:
                                day=datelist[0]
                                month=datelist[-1]
                        #if the date cell was empty, an error would appear that is caught before
                        except IndexError:
                            day="unclear"
                            month="unclear"
                        print(day, month)
                    if i==2:
                        attacktype=cell.text
                    if i==3:
                        deaths=cell.text
                    if i==4:
                        injured=cell.text
                    if i==5:
                        location=cell.text
                        loclist = location.split()
                        country=loclist[-1]
                    if i==6:
                        details=cell.text
                    if i==7:
                        attacker=cell.text
                    if i==8:
                        context=cell.text
                        #finally we add all retrieved elements to a list and write the list as a new row to the csv
                        output_list=(day,month,year,attacktype,deaths,injured,location,details,attacker,context,country,link)
                        resultsfileWriter.writerow(output_list)
```

```python
#covers for 1980 to 1986
localtime = datetime.now().strftime("%Y-%b-%d--%H-%M-%S")
with open(("terror_1980to_1986_{}.csv").format(localtime), 'w') as resultsfile:
    
    resultsfileWriter = csv.writer(resultsfile)
    firstrow=("day","month","year","attacktype","deaths","injured","location","details","attacker","context","country","link")
    resultsfileWriter.writerow(firstrow)
    
    for year in range(1980,1987):
        year=str(year)
        link=("https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_{}".format(year))
        print(link)
        page = requests.get(link).text
        soup = BS(page, 'html.parser')
        tables = soup.find_all('table', class_="wikitable sortable")
        #we retrieve the h2s as we have not one but several tables here and
        #it is the most reliable to retrieve the h2s, which record the months 
        h2s= soup.find_all('h2')

        j=0

        for table in tables:
            #to go through the h2s to retrieve the month, we use j as an index
            j+=1
            span=h2s[j].find('span')
            monthhead=(span.text)
            #print(monthhead)
            rows = table.find_all('tr')
            for row in rows:
                i=0
                cells=(row.find_all("td"))
                for cell in cells:
                    #print(cell)
                    i+=1
                    if i==1:
                        celldate=cell.text
                        datelist=celldate.split()
                        month=monthhead
                        day=datelist[0]
                        print(day, monthhead)
                    if i==2:
                        attacktype=cell.text
                    if i==3:
                        deaths=cell.text
                    if i==4:
                        injured=cell.text
                    if i==5:
                        location=cell.text
                        loclist = location.split()
                        country=loclist[-1]
                    if i==6:
                        details=cell.text
                    if i==7:
                        attacker=cell.text
                    if i==8:
                        context=cell.text
                        output_list=(day,month,year,attacktype,deaths,injured,location,details,attacker,context,country,link)
                        resultsfileWriter.writerow(output_list)
```

```python
#cover for 1995 to 2000
localtime = datetime.now().strftime("%Y-%b-%d--%H-%M-%S")
with open(("terror_1995to_2000_{}.csv").format(localtime), 'w') as resultsfile:
    
    resultsfileWriter = csv.writer(resultsfile)
    firstrow=("day","month","year","attacktype","deaths","injured","location","details","attacker","context","country","link")
    resultsfileWriter.writerow(firstrow)
    
    for year in range(1995,2001):
        year=str(year)
        link=("https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_{}".format(year))
        print(link)
        page = requests.get(link).text
        soup = BS(page, 'html.parser')
        tables = soup.find_all('table', class_="wikitable")
        h2s= soup.find_all('h2')

        j=1

        for table in tables:
            #print(table)
            j+=1
            span=h2s[j].find('span')
            monthhead=(span.text)
            month=monthhead
            #print(monthhead)
            rows = table.find_all('tr')
            for row in rows:
                i=0
                cells=(row.find_all("td"))
                for cell in cells:
                    #print(cell)
                    i+=1
                    if i==1:
                        celldate=cell.text
                        datelist=celldate.split()
                        day=datelist[-1]
                        month=monthhead
                        print(day, monthhead)
                    if i==2:
                        attacktype=cell.text
                    if i==3:
                        deaths=cell.text
                    if i==4:
                        injured=cell.text
                    if i==5:
                        location=cell.text
                        loclist = location.split()
                        country=loclist[-1]
                    if i==6:
                        details=cell.text
                    if i==7:
                        attacker=cell.text
                    if i==8:
                        context=cell.text
                        
                #this prevents empty cells from being added        
                if cells!=[]:
                    output_list=(day,monthhead,year,attacktype,deaths,injured,location,details,attacker,context,country,link)
                    #print(output_list)
                    resultsfileWriter.writerow(output_list)
```
