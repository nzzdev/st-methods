```python
# coding: utf-8

# In[65]:

from multiprocessing import Pool
from joblib import Parallel, delayed
import multiprocessing
import pandas as pd
import re




#data.head(12)


# In[66]:

# In[6]:
#lex = pd.read_csv(open("Nazi_Language4.csv",'rU'), encoding='utf-8', engine='c', dtype=str)

lex = pd.read_csv(open('/Users/halukamaier-borst/Documents/FTstuff/GermanElections/AfDComments/Step05DeployAgain/Nazi_Language4.csv','rU'), encoding='utf-8', engine='c', dtype=str, header=0, quotechar = '"')
lex=lex.sort_values(by="Phrase")
lex=lex.reset_index()
lex=lex.drop(["index"], axis=1)
#print(lex[0:5])
#print(list(lex))
#print(lex.head(30))

# In[67]:

#(data.head())


# In[68]:

# In[20]:

# In[7]:
#print(lex)
new_lex=lex
halfindex=int((len(new_lex.index))/2)
halfword=(new_lex["Phrase"][halfindex])
halfplusindex=(halfindex+1)
halfplusword=(new_lex["Phrase"][halfindex+1])

#print(halfindex)
#print(halfword)
#print(halfplusindex)
#print(halfplusword)

while(halfword[0]==halfplusword[0]):
    #print("define")
    halfindex+=1
    halfplusindex=halfindex+1
    halfword=(new_lex["Phrase"][halfindex])
    halfplusword=(new_lex["Phrase"][halfplusindex])



uphalfLex=new_lex.ix[:halfindex]
lowhalfLex=new_lex.ix[(halfplusindex):]
uphalfLex.reset_index(drop=True, inplace=True)
lowhalfLex.reset_index(drop=True, inplace=True)
print(uphalfLex)
#print(lowhalfLex)
```

```python
upquarterindex=int((len(uphalfLex.index))/2)
upquarterplusindex=upquarterindex+1
upquarterword=(uphalfLex["Phrase"][upquarterindex])
upquarterplusword=(uphalfLex["Phrase"][upquarterindex+1])
while (upquarterword[0]==upquarterplusword[0]):
    upquarterindex+=1
    upquarterplusindex=upquarterindex+1
    upquarterword=(uphalfLex["Phrase"][upquarterindex])
    upquarterplusword=(uphalfLex["Phrase"][upquarterplusindex])

#print(upquarterword)
#print(upquarterplusword)
upupquarterLex=uphalfLex.ix[:upquarterindex]
uplowquarterLex=uphalfLex.ix[upquarterplusindex:]

upupquarterLex.reset_index(drop=True, inplace=True)
uplowquarterLex.reset_index(drop=True, inplace=True)
#print(upupquarterLex)
print(uplowquarterLex)
```

```python
# In[59]:

lowquarterindex=int((len(lowhalfLex.index))/2)

lowquarterword=(lowhalfLex["Phrase"][lowquarterindex])
lowquarterplusindex=lowquarterindex+1

lowquarterplusword=(lowhalfLex["Phrase"][lowquarterindex+1])
while (lowquarterword[0]==lowquarterplusword[0]):
    lowquarterindex+=1
    lowquarterplusindex=lowquarterindex+1
    lowquarterword=(lowhalfLex["Phrase"][lowquarterindex])
    lowquarterplusword=(lowhalfLex["Phrase"][lowquarterplusindex])

lowupquarterLex=lowhalfLex.ix[:lowquarterindex]
lowlowquarterLex=lowhalfLex.ix[(lowquarterplusindex):]

lowupquarterLex.reset_index(drop=True, inplace=True)
lowlowquarterLex.reset_index(drop=True, inplace=True)
#print(lowupquarterLex)
print(lowlowquarterLex)
```

```python
# In[11]:
#cellvavlist=[]
#leftrightlist=[]
#midlist=[]
#for cell in (data['text']):
 #   for word in (cell.split()):
import re
data1 = AfdComments_drop

#[x.strip() for x in my_string.split(',')]
#rightlist=[]
#leftlist=[]
#data1=data

#for each cell in the text column of data

def decode(b):
    return bytes(b, 'utf-8').decode('unicode-escape').encode('latin1').decode('utf-8')


def lexiconclassifier(line):


    cell=(line['Naziwords'])
    comment_id=(line['comment_id'])
    comment_author=(line['Author_id'])
    comment_date=(line['date'])
    comment_likes=(line['likes'])
    wordcount=0
    cellrating=0
    wordlist=[]
    ratingfactor=0



    new_lex=pd.DataFrame()

    #lr_classifier=0
    cell=cell[1:-1]

    #for reach word in cell
    element={}

    cell=cell.replace(',', " ")
    for word in re.split(',| ',cell):
        
        wordcount+=1

        word=word.lower()
        word=word.replace("'", "")
        word=word.replace('"', "")
        
        #print(word)



        #iterates through the lexicon

        try:
            if word[0]<=halfplusword[0]:
                if word[0]<=upquarterplusword[0]:
                    new_lex=upupquarterLex
                else:
                    new_lex=uplowquarterLex
            else:
                if word[0]<=lowquarterplusword[0]:
                    new_lex=lowupquarterLex
                else:
                    new_lex=lowlowquarterLex
        except LookupError:
            pass

        for index, row in (new_lex).iterrows():
            #if word is found in the lexicon
            #print str.startswith( 'this' )
            if word.startswith(row["Phrase"]):
                wordcolumn=str(row["Phrase"])

                ratingfactor+=1
                #print("Nazi!")
                wordlist.append(word)
                element[wordcolumn]=1




    #print(ratingfactor)
    #print(comment_id)
    element['Naziwordcounts']=ratingfactor
    element['wordcount']=wordcount
    element['comment_id']=comment_id
    element['Author_id']=comment_author
    element['date']=comment_date
    element['likes']=comment_likes
    element['Naziwords']=str(wordlist)

    return(element)

num_cores = multiprocessing.cpu_count()
print(num_cores)

result=Parallel(n_jobs=num_cores)(delayed(lexiconclassifier)(line) for index,line in (data1).iterrows())




test=pd.DataFrame.from_dict(result, orient='columns', dtype=None)

test.head()
# In[73]:



# In[ ]:
```
