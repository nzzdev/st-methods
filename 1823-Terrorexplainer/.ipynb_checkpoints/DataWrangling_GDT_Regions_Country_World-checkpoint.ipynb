```python
import pandas as pd
import glob
```

```python
GDT=pd.read_csv("GDT_database.csv")
print(len(GDT))
```

```python
GDT
```

```python
GDTclean=GDT[(GDT["crit1"]&GDT["crit2"]&GDT["crit3"]==1)]
GDTclean=GDTclean[GDT["nkill"]!=0]
GDTclean=GDTclean[GDT["doubtterr"]!=1]

print(len(GDTclean))
GDTclean
```

```python
GDTclean["incidentno"]=1
GDTclean
```

```python
GDTregion=GDTclean.groupby(["region_txt","iyear"]).sum()
GDTcountry=GDTclean.groupby(["country_txt","iyear"]).sum()
GDTworld=GDTclean.groupby(["iyear"]).sum()
```

```python
GDTregion
```

```python
GDTregion=GDTregion.reset_index()
GDTcountry=GDTcountry.reset_index()
GDTworld=GDTworld.reset_index()
GDTregion
```

```python
pd.options.display.max_columns=100
pd.options.display.max_rows=10
GDTcleanYM=GDTworld
```

```python
yearbased=GDTcleanYM
yearbased["iyear"]=pd.to_datetime(yearbased["iyear"], format="%Y").dt.year
yearbased
yearbased.to_csv("yeartoyear.csv", index=False)
```
