```python
#Loading in the data with the three-year averages for dev aid and the numbers for emigration into OECD countries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from  matplotlib.ticker import FuncFormatter
migVSmoney=pd.read_csv("FinalData/MigVsMoney_Cont_Reg31.csv")
migVSmoney.head()
```

```python
#multiplying to adjust the metrics
migVSmoney["Report Year"]=migVSmoney["Report Year"].astype(int)
migVSmoney["Emigration per cap over all OECD"]=migVSmoney["Emigration per cap over all OECD"]*1000000
migVSmoney["3yrAverage Money"]=migVSmoney["3yrAverage Money"]/1000000

migVSmoney.head(20)
```

```python
#Now plot a timeline specifically for Ethiopia
recipient="Ethiopia"
data=migVSmoney[migVSmoney["Recipient"]==recipient]
fig = plt.figure(figsize=(16, 9), dpi=100)
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42

sns.set_style("whitegrid")


ax1 = fig.add_subplot(111)
x = data['Report Year'].astype(int).as_matrix(columns=None)
plt.xticks(np.arange(min(x), max(x), 2))
plt.xlabel('Jahr', fontsize=10)

y1=data['3yrAverage Money'].as_matrix(columns=None)
y2=data['Emigration per cap over all OECD'].as_matrix(columns=None)
#y1ticks = ['0','1000','2000','3000','4000','5000']
äy2ticks = ['0','100','200','300','400','500']

ax1.plot(x,y1,label='Entwicklungshilfe',)
plt.ylim(0,5000)
plt.legend(loc="lower left")
plt.ylabel('Entwicklungshilfe in Millionen US-Dollar (Durchschnitt über drei Jahre)', fontsize=10)


ax2 = ax1.twinx()
ax2.set_ylim(0, 500)

ax2.plot(x,y2, color="red", label='Migration')
plt.legend(loc="lower right")
plt.ylabel('Auswanderer pro tausend Einwohner', fontsize=10, rotation=270, labelpad=20)

ax2.set_yticks(np.linspace(ax2.get_yticks()[0],ax2.get_yticks()[-1],len(ax1.get_yticks())))
#ax1.set_yticks(np.linspace(ax1.get_yticks()[0],ax1.get_yticks()[-1],len(ax2.get_yticks())))
plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda x, _: int(x)))

plt.margins(x=0)

# ADD THIS LINE
#align_y_axis(ax1, ax2, 1, 1)
plt.xlabel('Years', fontsize=10)
#plt.yticks(y1,y1ticks) # This makes your desired y ticks
#plt.yticks(y2,y1ticks) # This makes your desired y ticks

plt.title("Äthiopien")
plt.savefig(recipient+'.pdf', transparent = True)

plt.show()
```
