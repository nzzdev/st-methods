```python
import pandas as pd
import glob
```

```python
GDT=pd.read_csv("GDT_database.csv")
print(len(GDT))
```

```python
GDT
```

```python
GDTclean=GDT[(GDT["crit1"]&GDT["crit2"]&GDT["crit3"]==1)]
GDTclean=GDTclean[GDT["nkill"]!=0]
print(len(GDTclean))
GDTclean
```

```python
GDTclean["incidentno"]=1
GDTclean
```

```python
GDTclean=GDTclean.groupby(["iyear"]).sum()
GDTclean
```

```python
GDTcleanYM=GDTclean.reset_index()
GDTcleanYM
```

```python
pd.options.display.max_columns=100
pd.options.display.max_rows=10
GDTcleanYM
```

```python
yearbased=GDTcleanYM
yearbased["date"]=pd.to_datetime(yearbased["iyear"], format="%Y")
yearbased
```

```python
yearbased["date"]=yearbased["date"].dt.year
```

```python
yearbased
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.figure(figsize=(12,9))
sns.set_style("ticks")
x=yearbased["iyear"]
y=yearbased["incidentno"]
plt.title("Anschläge pro Jahr")
plt.plot(x,y,marker='o')
plt.show()
```

```python
plt.figure(figsize=(12,9))
sns.set_style("whitegrid")
x=yearbased["iyear"]
y=(yearbased["nkill"])
plt.title("Tote durch Anschläge pro Jahr")

plt.plot(x,y,marker='o')
plt.show()
```

```python
yearbased.to_csv("yeartoyear.csv", index=False)
```
