```python
import pandas as pd
```

Filtering for our defined criteria

```python
GDT=pd.read_csv("../RawData/GDT_clean_database_2.csv")
print(len(GDT))
GDTclean=GDT[(GDT["crit1"]&GDT["crit2"]&GDT["crit3"]==1)]
GDTclean=GDTclean[GDTclean["nkill"]!=0]
GDTclean=GDTclean[GDTclean["doubtterr"]!=1]
GDTclean=GDTclean[GDTclean["gname"]!="Unknown"]

GDTclean["incidentno"]=1
GDT_reduced=GDTclean
print(len(GDTclean))
GDTclean
```

```python
classification=pd.read_csv("TerrorGroups_Over100Deaths_2017_2.csv", encoding="latin-1")
classification
```

Only taking in the safe cases where we have two sources that classify the terrorist group

```python
safecases=classification[(classification["source1"].notnull()) & (classification["source2"].notnull())]
safecases
```

Creating a dictionary of unique terrorgroup classifications

```python
classlist=[]
a=classification["classification1"].unique().tolist()
b=classification["classification2"].unique().tolist()
c=classification["classification3"].unique().tolist()
classlist.extend(a)
classlist.extend(b)
classlist.extend(c)
classlist = [classlist for classlist in classlist if str(classlist) != 'nan']
classlist=set(classlist)
```

Creating an empty dateframe in which we push our results

```python
wholedf=pd.DataFrame()
wholedf
```

Now we iterate through all the ideological classifications and sum over the years kills, wounds and incidents

```python
for groupclass in classlist:
    groupfilter=[groupclass]
    groupfilterclass=str(groupfilter).replace("[","").replace("]","")
    print(groupfilterclass)
    groupfilter=[groupfilterclass.strip("'")]
    print(groupfilter)
    #collecting the groups of a certain classification and all other groups as non.group 
    group=safecases[(safecases["classification1"].isin(groupfilter)) | (safecases["classification2"].isin(groupfilter))| (safecases["classification3"].isin(groupfilter))]
    nongroup=safecases[~((safecases["classification1"].isin(groupfilter)))&~((safecases["classification2"].isin(groupfilter)))&~((safecases["classification3"].isin(groupfilter)))]
    group 
    
    #filtering the cases by group and non-group
    grouplist=group["gname"].tolist()
    groupGTD=GDTclean[(GDTclean["gname"].isin(grouplist))]
    nongrouplist=nongroup["gname"].tolist()
    nongroupGTD=GDTclean[(GDTclean["gname"].isin(nongrouplist))]
    
    #then summing up the cases by year and reducing the dataframe to deaths, injured and number of incidents
    bygroupyear=groupGTD.groupby("iyear").sum()
    bygroupyear=bygroupyear.reset_index()
    bygroupyear=bygroupyear[["iyear","nkill","nwound","incidentno"]]

    bynongroupyear=nongroupGTD.groupby("iyear").sum()
    bynongroupyear=bynongroupyear.reset_index()
    bynongroupyear=bynongroupyear[["iyear","nkill","nwound","incidentno"]]
    
    #merging both and creating a dataframe that is written to a csv
    comparison=pd.merge(bygroupyear,bynongroupyear, how='outer', on="iyear", suffixes=(('_{0}'.format(groupfilter[0])), ('_non_{0}').format(groupfilter[0])))
    comparison=comparison.sort_values(by="iyear")
    comparison.to_csv("overtime_{0}.csv".format(groupfilter[0]))
    comparison
```
