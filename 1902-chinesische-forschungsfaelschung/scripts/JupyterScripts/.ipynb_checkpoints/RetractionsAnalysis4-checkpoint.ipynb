```python
import pandas as pd
```

```python
rawdata=pd.read_csv("../RawData/retractions.csv")
rawdata.head()
```

```python
rawdata["reason"]
```

```python
reasons=list(rawdata["reason"].unique())
authors=list(rawdata["scientist"].unique())
reasons
```

```python
for i in reasons:
    if(type(i)==float):
        print(i)
reasons = [x for x in reasons if str(x) != 'nan']
for i in authors:
    if(type(i)==float):
        print(i)
authors = [x for x in authors if str(x) != 'nan']
```

```python
reasonslist=[]
for i in reasons:
    reasonslist.extend(i.split(","))
reasonslist=list(filter(None, reasonslist))
reasonslist=list(set(reasonslist))
reasonslist

authorlist=[]
for i in authors:
    authorlist.extend(i.split(","))
authorlist=list(filter(None, authorlist))
#authorlist=list(set(authorlist))
authorlist
```

```python
pd.DataFrame({"reasons":reasonslist}).to_csv("listofreasons.csv", index=False)
```

```python
from collections import Counter
p=Counter(authorlist)
type(p)
```

```python
df = pd.DataFrame.from_dict(p, orient='index').reset_index()
df
```

```python
df=df.sort_values(by=0, ascending=False)
df.to_csv("researchers.csv", index=False)
df
```

```python
seriousoffences=['Duplication of Article','Duplication of Data','Duplication of Image','Ethical Violations by Author','Euphemisms for Duplication','Euphemisms for Plagiarism','Falsification/Fabrication of Data','Falsification/Fabrication of Image','Investigation by Company/Institution','Investigation by Journal/Publisher','Investigation by Third Party','Manipulation of Images','Misconduct - Official Investigation/Finding','Misconduct by Author','Plagiarism of Article','Plagiarism of Image','Plagiarism of Text']
```

```python
serious_df=pd.DataFrame()
for i in seriousoffences:
    serious_cases=rawdata[rawdata["reason"].str.contains(i, regex=False, na=False)]
    serious_df=pd.concat([serious_df,serious_cases])
```
