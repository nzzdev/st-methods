```python
import pandas as pd
```

```python
#read in the csv with all the raw data from retraction watch
rawdata=pd.read_csv("../RawData/retractions_raw.csv")
singlepapers=rawdata.drop_duplicates(subset="paper")
print(len(singlepapers))
```

The problem is that in some columns, we have multiple entries, so we have to reduce that complexity

```python
#we create three lists from this dataframe. one for reasons and one for authors and one for subjects. 
reasons=list(rawdata["reason"])
authors=list(rawdata["scientist"])
subjects=list(rawdata["subject"])
```

```python
#now we filter these three lists and excluded nans 
for i in reasons:
    if(type(i)==float):
        print(i)
reasons = [x for x in reasons if str(x) != 'nan']

for i in authors:
    if(type(i)==float):
        print(i)
authors = [x for x in authors if str(x) != 'nan']

for i in subjects:
    #print(i)
    if(type(i)==float):
        print(i)
subjects = [x for x in subjects if str(x) != 'nan']
subjects
```

```python
#then we want to split the reasons and the authors in the lists further up 
#because we have multiple entries in the same element of the list such as 'Duplication of Article,Error by Journal/Publisher,'

#we create an empty, clean list for all the reasons
reasonslist=[]

#we go through the messy list of reasons
for i in reasons:
    #we split each element of the messy list on commas and add these splitted elements to the clean list
    element=i.split(",")
    reasonslist.extend(element)

    
#then we filter for Nones, empty spaces in the list and leading and trailing spaces in strings
reasonslist=list(filter(None, reasonslist))
reasonslist=list(filter((" ").__ne__, reasonslist))
reasonslist = [x.strip() for x in reasonslist]


#and we reduce the list to unique entries
unique_reasonslist=list(set(reasonslist))
unique_reasonslist


#same here for authors
authorlist=[]
for i in authors:
    element=i.split(",")
    authorlist.extend(element)
authorlist=list(filter(None, authorlist))
authorlist=list(filter((" ").__ne__, authorlist))
authorlist = [x.strip() for x in authorlist]


unique_authorlist=list(set(authorlist))
len(unique_authorlist)


#same here for subjects
subjectlist=[]

for i in subjects:
    element=i.split("; ")
    subjectlist.extend(element)
subjectlist=list(filter(None, subjectlist))
subjectlist=list(filter((" ").__ne__, subjectlist))
subjectlist=list(filter((",").__ne__, subjectlist))

subjectlist = [x.strip() for x in subjectlist]

unique_subjectlist=list(set(subjectlist))
print(unique_subjectlist)
```

```python
#Now, we want to count for each author how often he can be found in the dataset and save the result as dictionary
authorsdict={}
a=0

#we go over the list of unique author names
for i in unique_authorlist:
    a+=1
    #we convert the current element of the list into a string
    variable_name=str(i)
    #search for this name in the dataframe and only keep rows that contain this name. also, we drop duplicate entries by searching for papers
    filtered_df=(rawdata[rawdata["scientist"].str.contains(variable_name, regex=False, na=False)]).drop_duplicates(subset="paper")
    #we read out the length of this dataframe
    len_authors=len(filtered_df)
    #then we save the name of the author and the length of the reduced dataframe 
    authorsdict[variable_name]=len_authors

#lastly we convert this dictionary to a dataframe, sort it and save it
authorsdf = pd.DataFrame.from_dict(authorsdict,orient='index')
authorsdf=authorsdf.sort_values(by=0, ascending=False)
authorsdf.to_csv("authors_by_occurence.csv")
authorsdf  
```

```python
#we do the exact same for reasons of retractions
reasonsdict={}
for i in unique_reasonslist:
    variable_name=str(i)
    filtered_df=(rawdata[rawdata["reason"].str.contains(variable_name, regex=False, na=False)]).drop_duplicates(subset="paper")
    len_reasons=len(filtered_df)
    reasonsdict[variable_name]=len_reasons

reasonsdf = pd.DataFrame.from_dict(reasonsdict,orient='index')
reasonsdf=reasonsdf.sort_values(by=0, ascending=False)
reasonsdf.to_csv("reasons_by_occurence.csv")
reasonsdf 
```

```python
#we do the exact same for subjects

subjectsdict={}
for i in unique_subjectlist:
    variable_name=str(i)
    filtered_df=(rawdata[rawdata["subject"].str.contains(variable_name, regex=False, na=False)]).drop_duplicates(subset="paper")
    len_subjects=len(filtered_df)
    subjectsdict[variable_name]=len_subjects

subjectsdf = pd.DataFrame.from_dict(subjectsdict,orient='index')
subjectsdf=subjectsdf.sort_values(by=0, ascending=False)
subjectsdf.to_csv("subjects_by_occurence.csv")
subjectsdf
```

now we want to filter for serious reasons for retractions that we defined previously

```python
#this is the list of these reasons
seriousoffences=[  'Concerns/Issues About Authorship','Concerns/Issues About Data','Concerns/Issues About Image','Concerns/Issues About Results','Concerns/Issues about Referencing/Attributions','Concerns/Issues about Third Party Involvement','Fake Peer Review', 'Conflict of Interest', 'Ethical Violations by Author', 'Euphemisms for Plagiarism','Falsification/Fabrication of Data','Falsification/Fabrication of Image','Falsification/Fabrication of Results','Manipulation of Images','Manipulation of Results','Misconduct by Author','Plagiarism of Article','Plagiarism of Data', 'Plagiarism of Image', 'Plagiarism of Text']
```

```python
#we filter the raw dataset only for rows which contain one of these serious reasons
df_filtered_for_serious_offences=rawdata[rawdata["reason"].str.contains('|'.join(seriousoffences), na=False)]
print(len(df_filtered_for_serious_offences))

#and throw away entries where we cannot use paper AND publishdoi to check if they are duplicates
indexNames = df_filtered_for_serious_offences[((df_filtered_for_serious_offences['paper'].isnull())|(df_filtered_for_serious_offences['paper'] == "Unavailable; ")) & ((df_filtered_for_serious_offences['publishdoi'].isnull())|(df_filtered_for_serious_offences['publishdoi'] == "unavailable")) ].index
serious_df=df_filtered_for_serious_offences.drop(indexNames)
print(len(serious_df))

#drop duplicates by filtering for paper names and doi
serious_df_reduced=serious_df.drop_duplicates(subset=["paper"]).drop_duplicates(subset=["publishdoi"])


#and end up with the length of the filtered dataframe
print(len(serious_df_reduced))
serious_df
```

```python
serious_df.to_csv("serious_offences_filtered.csv", index=False)
len(serious_df)
```

Why do we keep both dataframes, one with duplicates and one without? Well, counting the length of the one with duplicates would be overestimating the number of problematic papers. But if we only keep the one without duplicates, we might miss out some reasons/authors.

```python
serious_df["university"][2]
```

```python
#we read out the entire column about universities
unis=serious_df["university"]

#filter for strings that are not nan
unis = [x for x in unis if str(x) != 'nan']
#then create an empty list for countries
countrieslist=[]

#we take each element of this column
for i in unis:
    #split it at the semicolon because this is the sign used between two different university addresses such as:
    #Health Services Research, Norwich Medical School, University of East Anglia, Norwich, UK; Public Health, Thurrock Council, Grays, UK;""
    uni=i.split(";")
    for j in uni:
        #then we split each university address by commas and use always the last element because this where most likely the country is listed
        uniaddress=j.split(",")
        elements=(uniaddress[-1])
        #and add it to the countrieslist
        countrieslist.append(elements)
```

```python
#then we clean this list for nans etx.
countrieslist=list(filter(None, countrieslist))
countrieslist=list(filter((" ").__ne__, countrieslist))
countrieslist = [x.strip() for x in countrieslist]

#and make this into a unique list
unique_countrieslist=list(set(countrieslist))
unique_countrieslist.sort()
unique_countrieslist
```

```python
#now we go through this unique list and create a dataframe that lists up all the countries
countrydict={}
for i in unique_countrieslist:
    variable_name=str(i)
    filtered_df=(serious_df[serious_df["university"].str.contains(variable_name, regex=False, na=False)]).drop_duplicates(subset=["paper"]).drop_duplicates(subset=["publishdoi"])
    len_countries=len(filtered_df)
    countrydict[variable_name]=len_countries

countrydf = pd.DataFrame.from_dict(countrydict,orient='index')
countrydf=countrydf.sort_values(by=0, ascending=False)
countrydf.to_csv("countries_by_occurence_serious.csv")
countrydf  
```
