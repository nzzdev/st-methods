# Newsflash-Terror-DBAnalyser

This script was developed to put a terrorist attack into a first context and should ideally be used together with the "Newsflash-Terror-Scraper". It should allow statements like:

<i>"While X attacks happened in Iraq in 2016, it was only Y attacks in Switzerland for the same timespan". </i>

<i>"The number of attacks have continously risen in the Middle-East over the last 10 years". </i>

<i>"On average every two days an attack happened in Pakistan". </i>  


  
The database used for this is Global Terrorism Database (GTD), which is the gold-standard in social science about terrorism. 

<br>
<br>

<b>However, there is a catch.</b> The GTD is only updated once a year with a lag of approximately six months. So for example, as I am writing this script, the next version of this will be published in summer 2018 and it will be only about 2017. And the current version used here is the GTD with data from 1970 up till end of 2016. Hence, it is <b>NOT POSSIBLE</b> to make statements based on the data such as:
<br>

<b>Wrong:</b><i> Leading up to the most recent attack in Mogadishu, the number of attacks has been constantly rising for the past few months. </i>

<b>Wrong:</b><i> This is the Xth attack by the Islamic State this year.</i>

<br>

Also, please do not try to compare data from the scraper against the GTD because it will be a problematic method 9.5 times out of 10. Best thing to do is to not mix up the both data sources for any graphic.

## Year (Works a bit different, be careful!!!)
Lastly, you have to set the year-filter, which works a little different as all the others.
If you leave it at default with:
<i>yearfilter=[0]</i>
It will show everything for the entire timespan from 1970 to the latest year

If you set the yearfilter;
<i>yearfilter=[2011,2016]</i><br>
It will show all data for these two years, 2011 and 2016

And if you want to show all the data between two years, you have to write:<br><br>
<i>yearfilter=list(range(2011, 2016))</i>
It will show all data from 2011, 2012, 2013, 2014, 2015 <b>BUT NOT</b> for the year <i>2016</i>
<br>
<br>
<br>

So if you want to include even 2016, you have to write:
<br><br>
<i>yearfilter=list(range(2011, <b>2017</b>))</i>

```python
yearfilter=[2016]
print(yearfilter)
```

```python
subframe=GDTclean
criteria_txt_array=[]
criteria_array=[]


if yearfilter[0]!=0:
    subframe=subframe[subframe["year"].isin(yearfilter)]
    sub_array=[]
    sub_array.append("for the year(s):")
    sub_array.extend(yearfilter)
    criteria_txt_array.append(sub_array)
    criteria_array.append("year")
    
if regionsfilter[0]!="XYZ":
    subframe=subframe[subframe["region"].isin(regionsfilter)]
    sub_array=[]
    sub_array.append("for the region(s):")
    sub_array.extend(regionsfilter)
    criteria_txt_array.append(sub_array)
    criteria_array.append("region")


if countryfilter[0]!="XYZ":
    subframe=subframe[subframe["country"].isin(countryfilter)]
    sub_array=[]
    sub_array.append("for the country(ies):")
    sub_array.extend(countryfilter)
    criteria_txt_array.append(sub_array)
    criteria_array.append("country")

    
if attackerfilter[0]!="XYZ":
    subframe=subframe[subframe["gname"].isin(attackerfilter)]
    sub_array=[]
    sub_array.append("for the attackergroup(s):")
    sub_array.extend(attackerfilter)
    criteria_txt_array.append(sub_array)
    criteria_array.append("gname")


        
if attacktypefilter[0]!="XYZ":
    subframe=subframe[(subframe["attacktype1_txt"].isin(attacktypefilter)) | (subframe["attacktype2_txt"].isin(attacktypefilter))| (subframe["attacktype3_txt"].isin(attacktypefilter))]
    sub_array=[]
    sub_array.append("for the attacktype(s):")
    sub_array.extend(attacktypefilter)
    criteria_txt_array.append(sub_array)

    
if weaptypefilter[0]!="XYZ":
    subframe=subframe[(subframe["weaptype1_txt"].isin(weaptypefilter)) | (subframe["weaptype2_txt"].isin(weaptypefilter))| (subframe["weaptype3_txt"].isin(weaptypefilter))]
    sub_array=[]
    sub_array.append("for the weapontype(s):")
    sub_array.extend(weaptypefilter)
    criteria_txt_array.append(sub_array)

if targtypefilter[0]!="XYZ":
    subframe=subframe[(subframe["targtype1_txt"].isin(targtypefilter)) | (subframe["targtype2_txt"].isin(targtypefilter))| (subframe["targtype3_txt"].isin(targtypefilter))]
    sub_array=[]
    sub_array.append("for the targettype(s):")
    sub_array.extend(targtypefilter)
    criteria_txt_array.append(sub_array)
    
subframe=subframe.sort_values(by="Qdate", ascending=False)
subframe=subframe.reset_index(drop=True)
subframe
```

```python
text_file = open("GTDsorteddatasets/sorteddata_description.txt", "w")
text_file.write("This is data\n")
for i in criteria_txt_array:
    text_file.write(str(i)[1:-1]+"\n")
text_file.close()
    
```

## Sort and data output
In this part of the script, you can sort the output according to "deaths", "injured" or "casualties". Default setting is "deaths".
The script gives you four different csv-files then in the reduceddatasets-folder.

1. "GTD_incidents_sortedby_date.csv": Attacks chronologically ordered with all background information.
2. "GTD_days_summed_sortedby_date.csv": Days with attacks. The number of deaths, injured and casualties is summed for each day. Days are listed in chronological order.

```python
subframe.to_csv("GTDsorteddatasets/GTD_incidents_sortedby_date.csv", index=False)
criteria_array.append("Qdate")
sumoverdays=subframe.groupby(criteria_array).sum()
sumoverdays=sumoverdays.reset_index()
sumoverdays=sumoverdays.sort_values(by="Qdate", ascending=True)
sumoverdays=sumoverdays.reset_index(drop=True)
criteria_array.extend(["deaths","injured","casualties","incidentno"])
sumoverdays=sumoverdays[criteria_array]
sumoverdays.to_csv("GTDsorteddatasets/GTD_days_summed_sortedby_date.csv", index=False)
sumoverdays
```

```python
subframe=GDTclean
criteria_txt_array=[]
criteria_array=[]
print(regionsfilter)
  
if regionsfilter[0]!="XYZ":
    subframe=subframe[subframe["region"].isin(regionsfilter)]
    sub_array=[]
    sub_array.append("for the region(s):")
    sub_array.extend(regionsfilter)
    criteria_txt_array.append(sub_array)
    criteria_array.append("region")


if countryfilter[0]!="XYZ":
    subframe=subframe[subframe["country"].isin(countryfilter)]
    sub_array=[]
    sub_array.append("for the country(ies):")
    sub_array.extend(countryfilter)
    criteria_txt_array.append(sub_array)
    criteria_array.append("country")

    
if attackerfilter[0]!="XYZ":
    subframe=subframe[subframe["gname"].isin(attackerfilter)]
    sub_array=[]
    sub_array.append("for the attackergroup(s):")
    sub_array.extend(attackerfilter)
    criteria_txt_array.append(sub_array)
    criteria_array.append("gname")


        
if attacktypefilter[0]!="XYZ":
    subframe=subframe[(subframe["attacktype1_txt"].isin(attacktypefilter)) | (subframe["attacktype2_txt"].isin(attacktypefilter))| (subframe["attacktype3_txt"].isin(attacktypefilter))]
    sub_array=[]
    sub_array.append("for the attacktype(s):")
    sub_array.extend(attacktypefilter)
    criteria_txt_array.append(sub_array)

    
if weaptypefilter[0]!="XYZ":
    subframe=subframe[(subframe["weaptype1_txt"].isin(weaptypefilter)) | (subframe["weaptype2_txt"].isin(weaptypefilter))| (subframe["weaptype3_txt"].isin(weaptypefilter))]
    sub_array=[]
    sub_array.append("for the weapontype(s):")
    sub_array.extend(weaptypefilter)
    criteria_txt_array.append(sub_array)

if targtypefilter[0]!="XYZ":
    subframe=subframe[(subframe["targtype1_txt"].isin(targtypefilter)) | (subframe["targtype2_txt"].isin(targtypefilter))| (subframe["targtype3_txt"].isin(targtypefilter))]
    sub_array=[]
    sub_array.append("for the targettype(s):")
    sub_array.extend(targtypefilter)
    criteria_txt_array.append(sub_array)
    
subframe=subframe.sort_values(by="Qdate", ascending=False)
subframe=subframe.reset_index(drop=True)
subframe.head()
```

```python
subframe=subframe.sort_values(by="Qdate", ascending=False)
subframe=subframe.reset_index(drop=True)
criteria_array.append("year")
yearbased=subframe.groupby(criteria_array).sum()
yearbased=yearbased.reset_index()
yearbased
```

```python
try: 
    countrylist=yearbased["country"].unique()
    for country in countrylist:
            countryyearbased=yearbased[yearbased["country"]==country]
            yearbasedbar=countryyearbased[["country","year","incidentno"]]
            yearbasedbar.to_csv(("chartdata/incidents_barchart/incidentsperyear_barchart_{}.csv").format(country))
            text_file = open(("chartdata/incidents_barchart/incidents_barchart_description_{}.txt").format(country), "w")
            text_file.write(("This is data {}\n").format(country))
            for i in criteria_txt_array:
                text_file.write(str(i)[1:-1]+"\n")
            text_file.close()

            yearbasedstackedbar=countryyearbased[["country","year","deaths","injured"]]
            yearbasedstackedbar.to_csv(("chartdata/deaths_injured_stackbar/deathsandinjuredperyear_stacked_barchart_{}.csv").format(country))
            text_file = open(("chartdata/deaths_injured_stackbar/deaths_injured_stackbar_description_{}.txt").format(country), "w")
            text_file.write(("This is data{}\n").format(country))
            for i in criteria_txt_array:
                text_file.write(str(i)[1:-1]+"\n")
            text_file.close()

            yearbasedstackedbar
    
        
except KeyError:
    yearbasedbar=yearbased[["year","incidentno"]]
    yearbasedbar.to_csv("chartdata/incidents_barchart/incidentsperyear_barchart.csv")
    text_file = open("chartdata/incidents_barchart/incidents_barchart_description.txt", "w")
    text_file.write("This is data\n")
    for i in criteria_txt_array:
        text_file.write(str(i)[1:-1]+"\n")
    text_file.close()

    yearbasedstackedbar=yearbased[["year","deaths","injured"]]
    yearbasedstackedbar.to_csv("chartdata/deaths_injured_stackbar/deathsandinjuredperyear_stacked_barchart.csv")
    text_file = open("chartdata/deaths_injured_stackbar/deaths_injured_stackbar_description.txt", "w")
    text_file.write("This is data\n")
    for i in criteria_txt_array:
        text_file.write(str(i)[1:-1]+"\n")
    text_file.close()

    yearbasedstackedbar
```
